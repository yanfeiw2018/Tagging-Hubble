```python
from bs4 import BeautifulSoup
import re
```

```python
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome('/Users/yanfeiwang/Desktop2/Tagging-Hubble/chromedriver') 
```

```python
urls = {}
%store -r id_list
for i in id_list:
    urls[i] = 'http://archive.stsci.edu/proposal_search.php?mission=hst&id='+i
```

```python
urls
```

```python
prog_cat = {}
for i in id_list:
    driver.get(urls[i])
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    doc = soup.text
    doc1 = re.findall('\nProposal type: .*\n', doc)
    prog_cat[i] = doc1[0]
```
